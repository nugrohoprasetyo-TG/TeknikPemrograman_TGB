```python
from abc import ABC, abstractmethod

class Survey(ABC):
    def __init__(self, id_survey, lokasi, jenis):
        self.id_survey = id_survey
        self.lokasi = lokasi
        self.jenis = jenis 
    def info(self):
        print(f"ID Survey   : {self.id_survey}")
        print(f"Lokasi      : {self.lokasi}")
        print(f"Jenis       : {self.jenis}")

    @abstractmethod
    def processing(self):
        pass

class SeismicSurvey(Survey):
    def __init__(self, id_survey, lokasi, jenis, metode):
        super().__init__(id_survey, lokasi, jenis)
        self.metode = metode

    def processing(self):
        return f"Processing Seismic Survey dengan metode {self.metode} filtering"
    
class GravitySurvey(Survey):
    def __init__(self, id_survey, lokasi, jenis, resolusi):
        super().__init__(id_survey, lokasi, jenis)
        self.resolusi = resolusi

    def processing(self):
        return f"Processing Seismic Survey dengan resolusi {self.resolusi} upward continue"
    
class MagneticSurvey(Survey):
    def __init__(self, id_survey, lokasi, jenis, instrumen):
        super().__init__(id_survey, lokasi, jenis)
        self.intstrumen = instrumen

    def processing(self):
        return f"Processing Seismic Survey dengan instrumen {self.intstrumen} anomaly mapping"
    

surveys = [
    SeismicSurvey("S001", "Sumatera", "Seismik", "Filtering"),
    GravitySurvey("G001", "Jawa", "Gravitasi", "High"),
    MagneticSurvey("M001", "Kalimantan", "Magnetik", "Proton Magnetometer")
]

for s in surveys:
    s.info()
    print(s.processing())
    print("-"*40)
```

```python
with open("data.txt", "w") as f :
    f.write("shot,receiver,twt\n")
    f.write("0,100,200\n")
    f.write("0,200,350\n")
    f.write("0,300,450\n")
    f.write("0,400,600\n")

with open("data.txt", "r") as f :
    lines = f.readlines()
    n_twt = []
    shot = []
    receiver = []
    twt = []
    for line in lines :
        print(line.strip())
        if line.startswith("0"):
            parts = line.split(",")
            shot.append(int(parts[0]))
            receiver.append(int(parts[1]))
            twt.append(int(parts[2]))
for i in range(len(twt)) :
    n_twt.append((twt[i]-min(twt))/(600-min(twt)))
print("n_twt =", n_twt)
with open("data_normalized.txt", "w") as f :
    f.write("shot,receiver,twt,n_twt\n")
    for i in range(len(n_twt)) :
        f.write(f"{shot[i]},{receiver[i]},{twt[i]},{n_twt[i]:.4f}\n")
```
